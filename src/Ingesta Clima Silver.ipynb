# Ingesta en la Capa Silver

```python
#Se importan las librerias

from pyspark.sql.functions import col
```

```python
# se lee la tabla de la capa de bronze

df = spark.read.table("Weather.bronze.bronze_weather")
```

```python
# Se eliminan los duplicados y los nulos

df = df.dropDuplicates()
df = df.dropna()
```

```python
# Se castean los tipos de datos de cada columna del Dataframe

df = df.select(
col("location").cast("string"),
col("date").cast("date"),
col("precipitation").cast("double"),
col("temp_max").cast("double"),
col("temp_min").cast("double"),
col("wind").cast("double"),
col("weather").cast("string"),
col("ingestion_date").cast("timestamp")
)
```

```python
# se filtan los datos irrelevantes

df = df.filter(
    (col("temp_max") <= 60) &
    (col("temp_min") >= -50) &
    (col("precipitation") >= 0) &
    (col("wind") >= 0)
)
```

```python
# Se escribe el Dataframe en la capa de silver

df.write.mode("overwrite").saveAsTable("Weather.silver.silver_weather")
```
