# Ingesta a la Capa Bronze

```python
#Se importan las librerias
from pyspark.sql.functions import current_timestamp
```

```python
# se lee el archivo csv y se agrega la columna ingestion_date
df = spark.read.load("/Volumes/filestore/tables/data/weather.csv",format ="csv",header=True)
df = df.withColumn("ingestion_date",current_timestamp())
```

```python
#Se escribe el dataframe en la tabla
df.write.mode("overwrite").saveAsTable("bronze_dev.weather.bronze_weather")
```
